```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
df = pd.read_csv("https://drive.google.com/u/1/uc?id=1sV0JD4aSx0brbH_Jmz3ibrjdLhPsepGn&export=download")
df
```

```python
#Calculo o coeficiente de correlação
cc1 = df['Prova 1'].corr(df['Prova 2']) 
cc2 = df['Prova 1'].corr(df['Trabalho']) 
cc3 = df['Prova 2'].corr(df['Trabalho']) 
print("Coeficiente de correlação entre prova 1 e prova 2:", "%.2f" %cc1)
print("Coeficiente de correlação entre prova 1 e trabalho:", "%.2f" %cc2)
print("Coeficiente de correlação entre prova 2 e trabalho:","%.2f" % cc3)
```

O par mais fortemente relacionado é o entre prova 1 e prova 2

```python
#Imprimir Prova 1 e imprimir coeficiente entre trabalho com duas casas decimais
print(df [["Prova 1"]])
cc22 = "%.2f" % cc2
print('Coeficiente entre Prova 1 e Trabalho:', cc22)
#Definindo o que pode ser concluido com o valor obtido
if cc22 == '0.39':
  print('O valor obtido mostra que existe uma correlação fraca')


```

```python
#Valores de a e b para prova 1 e prova 2
(a, b) = np.polyfit(x=df["Prova 1"], y=df["Prova 2"], deg = 1)
print(a,b)
#Com três casas decimais
(aa,bb) = round(a,3),round(b,3)
print("Com três casas decimais:",aa,bb)
```

```python
#Imprimir reta dde regressão com a e b até três casas decimais
x = np.arange(0,51)
y = aa*x+bb
plt.plot(x,y,'r')
```
